# Joining table Corier_comapany_invoice and Company_X_Order_Report  // naming the table cci_cxor

```python
cci_cxor = Courier_company_invoice.merge(Company_X_Order_Report, left_on = 'Order ID', right_on = 'ExternOrderNo')
```

```python
cci_cxor.head()
```

# Joining table cci_cxor and table Company_X_SKU_Master 

```python
final_df = cci_cxor.merge(Company_X_SKU_Master, on = 'SKU')
```

```python
final_df.head()
```

# Converting gram to kg and making a new column as Total weight as per X (KG)

```python
final_df['Total weight as per X (KG)'] = final_df['Weight (g)']*final_df['Order Qty'] /1000
final_df.head()
```

# Adding  Weight slab as per x (KG) column and rounding the values to multiple of 0.5

```python
final_df['Weight slab as per X (KG)'] = final_df['Total weight as per X (KG)'].apply(lambda x: math.ceil(x*2) / 2)
final_df.head()
```

# creating a column as 'Total weight as per Courier Company (KG)'

```python
final_df['Total weight as per Courier Company (KG)'] = final_df['Charged Weight']
final_df.head()
```

# Creating 'Weight slab charged by Courier Company (KG)' column and rounding its values to multiple of 0.5

```python
final_df['Weight slab charged by Courier Company (KG)'] = final_df['Total weight as per Courier Company (KG)'].apply(lambda x: math.ceil(x*2) / 2)
final_df.head()
```

```python
final_df['Delivery Zone as per X'] = final_df['Zone'].apply(lambda x : x.capitalize())
final_df.head()
```

```python

final_df['Delivery Zone charged by Courier Company'] = final_df['Zone'].apply(lambda x : x.capitalize())
```

```python

final_df['Charges Billed by Courier Company (Rs.)'] = final_df['Billing Amount (Rs.)']
final_df.head()
```

# Selecting required column and saving into a new data frame as 'company_X'

```python
company_X = final_df[['Order ID','AWB Code','Total weight as per X (KG)','Total weight as per Courier Company (KG)','Weight slab charged by Courier Company (KG)','Delivery Zone as per X','Delivery Zone charged by Courier Company','Charges Billed by Courier Company (Rs.)','Type of Shipment']]
```

```python
company_X.head()
```

```python
Grouping data by 'Order ID' and creating a new data frame as 'company_x_new'
```

```python
company_x_new = company_X.groupby(['Order ID']).agg({'AWB Code': 'first','Total weight as per X (KG)': 'sum', 'Total weight as per Courier Company (KG)':'first'
                                     ,'Weight slab charged by Courier Company (KG)':'first' ,'Delivery Zone as per X':'first'
                                      ,'Delivery Zone charged by Courier Company':'first'
                                      ,'Charges Billed by Courier Company (Rs.)':'first'
                                      , 'Type of Shipment':'first'
                                     })
```

```python
company_x_new.head()
```

```python
company_x_new['Weight slab as per X (KG)'] = company_x_new['Total weight as per X (KG)'].apply(lambda x: math.ceil(x*2) / 2)
company_x_new.head()
```

# Creating Function for Expected Charge as per X (Rs.) Calculation

```python

def expected_charge(ty_smt,ws,zone):
    
    fwd_b_fixed = 33
     
    fwd_b_additional = 28.3
    
    rto_b_fixed = 20.5
    
    rto_b_additional = 28.3
    
    fwd_d_fixed = 45.4

    fwd_d_additional = 44.8

    rto_d_fixed = 41.3
    
    rto_d_additional = 44.8
    
    fwd_e_fixed = 56.6
    
    fwd_e_additional = 55.5
    
    rto_e_fixed = 50.7
    
    rto_e_additional = 55.5
    
    if ty_smt == 'Forward charges':
        if ws / .5 > 1:
            if zone == 'D':
                return fwd_d_fixed + (ws // .5 - 1)* fwd_d_additional
            elif zone == 'B':
                return fwd_b_fixed + (ws // .5 - 1)* fwd_b_additional
            else:
                return fwd_e_fixed + (ws // .5 - 1)* fwd_e_additional
        else:
            if zone == 'D':
                return fwd_d_fixed 
            elif zone == 'B':
                return fwd_b_fixed 
            else:
                return fwd_e_fixed 
    elif ty_smt == 'Forward and RTO charges':
        if ws / .5 > 1:
            if zone == 'B':
                return (fwd_b_fixed + rto_b_fixed) + (ws // .5 - 1)* (fwd_b_additional + rto_b_additional)
            elif zone == 'D':
                return (fwd_d_fixed + rto_d_fixed) + (ws // .5 - 1)* (fwd_d_additional + rto_d_additional)
            else:
                return (fwd_e_fixed + rto_e_fixed) + (ws // .5 - 1)* (fwd_e_additional + rto_e_additional)
        else:
            if zone == 'B':
                 return (fwd_b_fixed + rto_b_fixed)
            elif zone == 'D':
                 return (fwd_d_fixed + rto_d_fixed)
            else:
                 return (fwd_e_fixed + rto_e_fixed)
    else:
        return 1
    
    
```

```python
company_x_new['Expected Charge as per X (Rs.)'] = company_x_new.apply(lambda x: expected_charge(x['Type of Shipment'], x['Weight slab as per X (KG)'], x['Delivery Zone as per X']), 
                        axis=1)
    
```

```python
company_x_new.head()
```

```python
company_x_new['Difference Between Expected Charges and Billed Charges (Rs.)'] = round(company_x_new['Expected Charge as per X (Rs.)'] - company_x_new['Charges Billed by Courier Company (Rs.)'],2)
```
