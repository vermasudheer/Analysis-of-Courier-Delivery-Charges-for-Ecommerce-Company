# Converting gram to kg and adding a new column as Total weight as per X (KG)

```python
combined_df['Total weight as per X (KG)'] = combined_df['Weight (g)']*combined_df['Order Qty'] /1000
combined_df.head()
```

# Adding  Weight slab as per x (KG) column and rounding the values to  the multiple of 0.5

```python
combined_df['Weight slab as per X (KG)'] = combined_df['Total weight as per X (KG)'].apply(lambda x: math.ceil(x*2) / 2)
combined_df.head()
```

# Adding  a column as 'Total weight as per Courier Company (KG)'

```python
combined_df['Total weight as per Courier Company (KG)'] = combined_df['Charged Weight']
combined_df.head()
```

# Adding 'Weight slab charged by Courier Company (KG)' column and rounding its values to the multiple of 0.5

```python
combined_df['Weight slab charged by Courier Company (KG)'] = combined_df['Total weight as per Courier Company (KG)'].apply(lambda x: math.ceil(x*2) / 2)
combined_df.head()
```

```python
combined_df['Delivery Zone as per X'] = combined_df['Zone'].apply(lambda x : x.capitalize())
combined_df.head()
```

```python

combined_df['Delivery Zone charged by Courier Company'] = combined_df['Zone'].apply(lambda x : x.capitalize())
```

```python

combined_df['Charges Billed by Courier Company (Rs.)'] = combined_df['Billing Amount (Rs.)']
combined_df.head()
```

# Grouping data by 'Order ID' and creating a new data frame as 'final_df'

```python
final_df = company_X_df.groupby(['Order ID']).agg({'AWB Code': 'first','Total weight as per X (KG)': 'sum', 'Total weight as per Courier Company (KG)':'first'
                                     ,'Weight slab charged by Courier Company (KG)':'first' ,'Delivery Zone as per X':'first'
                                      ,'Delivery Zone charged by Courier Company':'first'
                                      ,'Charges Billed by Courier Company (Rs.)':'first'
                                      , 'Type of Shipment':'first'
                                     })
```

```python
final_df.head()
```

```python
final_df['Weight slab as per X (KG)'] = final_df['Total weight as per X (KG)'].apply(lambda x: math.ceil(x*2) / 2)
final_df.head()
```

# Expected Charge as per X (Rs.) Calculation


```python

def expected_charge(ty_smt,ws,zone):
    
    fwd_b_fixed = 33
     
    fwd_b_additional = 28.3
    
    rto_b_fixed = 20.5
    
    rto_b_additional = 28.3
    
    fwd_d_fixed = 45.4

    fwd_d_additional = 44.8

    rto_d_fixed = 41.3
    
    rto_d_additional = 44.8
    
    fwd_e_fixed = 56.6
    
    fwd_e_additional = 55.5
    
    rto_e_fixed = 50.7
    
    rto_e_additional = 55.5
    
    if ty_smt == 'Forward charges':
        if ws / .5 > 1:
            if zone == 'D':
                return fwd_d_fixed + (ws // .5 - 1)* fwd_d_additional
            elif zone == 'B':
                return fwd_b_fixed + (ws // .5 - 1)* fwd_b_additional
            else:
                return fwd_e_fixed + (ws // .5 - 1)* fwd_e_additional
        else:
            if zone == 'D':
                return fwd_d_fixed 
            elif zone == 'B':
                return fwd_b_fixed 
            else:
                return fwd_e_fixed 
    elif ty_smt == 'Forward and RTO charges':
        if ws / .5 > 1:
            if zone == 'B':
                return (fwd_b_fixed + rto_b_fixed) + (ws // .5 - 1)* (fwd_b_additional + rto_b_additional)
            elif zone == 'D':
                return (fwd_d_fixed + rto_d_fixed) + (ws // .5 - 1)* (fwd_d_additional + rto_d_additional)
            else:
                return (fwd_e_fixed + rto_e_fixed) + (ws // .5 - 1)* (fwd_e_additional + rto_e_additional)
        else:
            if zone == 'B':
                 return (fwd_b_fixed + rto_b_fixed)
            elif zone == 'D':
                 return (fwd_d_fixed + rto_d_fixed)
            else:
                 return (fwd_e_fixed + rto_e_fixed)
    else:
        return 1
    
    
```

```python
final_df['Expected Charge as per X (Rs.)'] = final_df.apply(lambda x: expected_charge(x['Type of Shipment'], x['Weight slab as per X (KG)'], x['Delivery Zone as per X']), 
                        axis=1)
    
```
